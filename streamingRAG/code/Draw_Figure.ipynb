```python
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots

plt.style.use('science')
# 上传并读取 CSV 文件
df1 = pd.read_csv('NaiveRAG_result.csv')
df2 = pd.read_csv('kmeans_RAG_result.csv')
df3 = pd.read_csv('streamingRAG_result.csv')
df4 = pd.read_csv('SKR_result.csv')

# 绘制合并后的曲线
plt.figure(figsize=(10, 6))

plt.plot(df1['K'], df1['Accuracy'], marker='o', label='NaiveRAG')
plt.plot(df2['K'], df2['Accuracy'], marker='o', label='kmeansRAG')
plt.plot(df3['K'], df3['Accuracy'], marker='o', label='streamingRAG')
plt.plot(df4['K'], df4['Accuracy'], marker='o', label='streaming_kmeans_RAG')

plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. K for Different Models')
plt.legend()
plt.grid()
plt.show()
```
